# Load Data and Check

```python
import pandas as pd

# Load data
df = pd.read_csv('exercise_final_with_equipment2.csv')

# Tampilkan kolom-kolom yang ada
print("Kolom yang tersedia:")
print(df.columns.tolist())

# Tampilkan jumlah baris dan kolom
print(f"\nJumlah baris: {df.shape[0]}")
print(f"Jumlah kolom: {df.shape[1]}")

# Tampilkan jumlah data kosong di setiap kolom
print("\nJumlah data kosong (missing values) per kolom:")
print(df.isnull().sum())
```

# Unique Value Check

```python
# Sajikan jumlah masing-masing nilai 'exercise_type' dalam tabel
type_counts = df['exercise_type'].value_counts().reset_index()
type_counts.columns = ['exercise_type', 'count']
print("Exercise Type")
display(type_counts)

# Sajikan jumlah masing-masing nilai 'body_part' dalam tabel
body_part_counts = df['body_part'].value_counts().reset_index()
body_part_counts.columns = ['body_part', 'count']
print("Body Part")
display(body_part_counts)

# Sajikan jumlah masing-masing nilai 'equipment' dalam tabel
equipment_counts = df['equipment'].value_counts().reset_index()
equipment_counts.columns = ['equipment', 'count']
print("Equipment")
display(equipment_counts)
```
